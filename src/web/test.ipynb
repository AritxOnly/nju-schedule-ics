```python
import requests
from IPython.display import Image
import base64
```

```python
response=requests.get("http://localhost:8999/get_login_session")

sess=response.json()
print(sess)
session=sess['session']
Image(data=base64.b64decode(sess['captcha_content']),format='png')
```

```python
response=requests.post(
    "http://localhost:8999/login",
    data={
        'session':session,
        'username':'PutYourOwnHere',
        'password':'Hahaha',
        'captcha_answer':'1e4f',
    }
)
print(response.text)
info=response.json()
print(info)
```

```python
url=f"http://localhost:8999/{info['uuid']}/schedule.ics"
response=requests.get(url)
print(url)
print(response.text)
```
